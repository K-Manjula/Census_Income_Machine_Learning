### TASK 3: Generate Hypothesis: Define Hypothesis statements

```python
To generate hypothesis statements, you need to define a null hypothesis (H₀) 
and an alternative hypothesis (H₁). These are foundational to statistical testing.
 step-by-step  using a dataset with income, education, gender, and race:
1. Impact of Education on Income
H₀ (Null Hypothesis): There is no significant relationship between a person's education level and their income.
H₁ (Alternative Hypothesis): There is a significant relationship between a person's education level and their income.

2. Impact of Gender on Income
H₀: Gender does not affect whether a person earns more than 50K.
H₁: Gender has a significant effect on whether a person earns more than 50K.

3. Impact of Race on Income
H₀: Income levels are independent of race.
H₁: Income levels vary significantly across different racial groups.

4. Effect of Weekly Working Hours on Income
H₀: Hours worked per week has no significant effect on income level.
H₁: Hours worked per week significantly affects income level.

General Template for Hypothesis
H₀: [No effect / no difference / no relationship]
H₁: [There is an effect / difference / relationship]
```

1. What is Hypothesis Testing? 
Hypothesis testing is a statistical method used to make decisions or inferences about population parameters based on sample data.
We always define:
Null Hypothesis (H₀): The statement we are trying to disprove (e.g., no effect, no difference).
Alternative Hypothesis (H₁): The statement we want to test for (e.g., there is an effect, a difference).

so for each feature that we explore, we define null hypothesis and alternative hypothesis.

The second step of hypothesis proposal and testing is to identify the skewness of the data. Depending on if the data has a normal distribution, is left skewed or right skewed, the methods to test the hypothesis will differ.


3. What is a p-value?
The p-value measures the probability of observing the data (or something more extreme) if the null hypothesis is true.
A small p-value (typically < 0.05) indicates strong evidence against H₀, so we reject H₀.

4. Type I and Type II Errors
Error Type	Description
Type I Error	Rejecting H₀ when it is actually true (false positive)
Type II Error	Not rejecting H₀ when H₁ is actually true (false negative)

```python
1.Age:
Null hypothesis - Age does not significantly impact income.
Alternative Hypothesis - Age impacts income.
to test the hypothesis we split the data frame into two groups - thos with high income and those with low income and take the ages of both those groups. 
We will then  calculate the t-value and p-value to check if the null hypothesis can be rejected in favour of alternative hypothesis.If
 t-value is negative, this means there is a significant difference in mean between the high income and low income groups.
 p-value is <0.05 means that the probability of this happening by chance is very low.
therefore, if the above two conditions are met, we can reject the null hypothesis and conclude that age has an impact on income earned.
Before performing chi-square test, we need to standerdize all the recordsin the income to <=50k and 50k and remove periods and spaces.
```

```python
#T-test: testing if average income differs between two versions to perform t-test we need numerical values
from scipy.stats import ttest_ind
import pandas as pd
# Assume df has income and group columns
# 'group' column contains 'A' and 'B'
group_A=df[df['income']=='<=50K']['age']
group_B=df[df['income']=='>50K']['age']
# Perform independent t-test
t_stat,p_value = ttest_ind(group_A, group_B,equal_var=False) # welch's t-test
print(f"T-statstic: {t_stat:.4f}")
print(f"p-value: {p_value:.4f}")
if p_value<0.05:
    print("Reject H0: There is a significant difference in  age between income groups")
else:
    print("Fail to Reject H0: No significant difference in age between  income groups")
```

```python
from the t-value and p-value, we can conclude that there is a statstically significant difference in age between 
the high and low income
groups and that income and age does have an impact on income. Null hypothesis can be rejected and the alternative hypothesis can be accepted.
we can also create a boxplot visualization to see the difference in age for the different income groups.
```

```python
bins=list(range(10,101,10)) # adjust range based on data
labels=[f"{i}-{i+9}" for i in bins[:-1]]
df['age_bucket']=pd.cut(df['age'],bins=bins, labels=labels, right=False)
grouped= df.groupby(['age_bucket', 'income']).size().unstack().fillna(0)
# convert counts to percentages
percentages= grouped.div(grouped.sum(axis=1),axis=0)*100
#plot
ax=percentages.plot(kind='bar',stacked=True, figsize=(10,6), color=['skyblue','salmon'])
plt.title('percentage of income groups by age bucket')
plt.xlabel('age bucket')
plt.ylabel('percentage of indiciduals')
plt.legend(title='income')

# Show plot
plt.show()
```

```python
3.sex
Null hypothesis -sex does not significantly impact income.
Alternative Hypothesis - sex impacts income.

income: categorical (<=50K, >50K)
sex: categorical (Male, Female)
to test the hypothesis , we perform the chi-square test. 
```

```python
import pandas as pd
from scipy.stats import chi2_contingency
#Create a contingency table
contingency = pd.crosstab(df['sex'], df['income'])

# Run Chi-Square Test
chi2, p, dof, expected = chi2_contingency(contingency)

# Results
print(f"Chi-Square Statistic: {chi2}")
print(f"P-value: {p}")

# Interpret the result
if p < 0.05:
    print("Reject the null hypothesis: Gender affects income.")
else:
    print("Fail to reject the null hypothesis: No significant effect of gender on income.")
```

```python
since the p value <0.05, we can conclude  that there is stastically significant association between sex and income.
we can reject the null hypothesis in favour of alternative hypothesis.
```
